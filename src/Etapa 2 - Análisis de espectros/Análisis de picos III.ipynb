Se toma como caso de prueba la imagen C1265-HD136488.

```python
import sys
sys.path.append("..") #Es necesario para importar Recursos.py

import Recursos

dirImg = Recursos.obtenerDirImagenes()
nombreImg = 'C1265-HD136488.fits'
resp = Recursos.graficarPicos(dirImg,nombreImg, 2) # cant_espectros = 2 
```

```python
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np 

hdu_list = fits.open(dirImg+'\\'+nombreImg, ignore_missing_end=True)
datos = np.apply_along_axis(sum, 0, hdu_list[0].data)  #se sumariza las columnas para graficar 
datos_norm = Recursos.normalize_MinMax(datos)
plt.plot(datos_norm)
```

```python
from scipy.signal import savgol_filter

datos_filter = savgol_filter(datos_norm, 61, 3, mode="nearest")
plt.plot(datos_filter)
```

```python
from scipy.signal import savgol_filter
from scipy.signal import find_peaks
#datos_filter2 = savgol_filter(datos_norm, 61, 3)
#datos_filter3 = savgol_filter(datos_norm,51,2)
datos_filter4 = savgol_filter(datos_norm,111 ,3)  #window_lenth --> impar
plt.plot(datos_norm, color='blue')
#plt.plot(datos_filter2, color='red')
#plt.plot(datos_filter3, color='green')
plt.plot(datos_filter4, color='red') #<701 order=5
peaks, properties = find_peaks(datos_filter4, prominence=0.4, width=30)
plt.plot(peaks, datos_filter4[peaks], "x", color='black')
plt.show()
```

```python
datos_filter4 = savgol_filter(datos_norm,91 ,2)  
plt.plot(datos_norm, color='blue')
plt.plot(datos_filter4, color='red')
plt.show()
```
