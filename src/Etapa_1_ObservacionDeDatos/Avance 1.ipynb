```python
import os
import numpy
import matplotlib.pyplot as plt
#%matplotlib inline
from skimage import io

elem_x_fila = 3

def actualizar(f,c):
    c += 1 
    if (c == elem_x_fila):
        f += 1
        c = 0
    return f,c

arch_png = []
for dirname, _, filenames in os.walk('png/'):
    for filename in filenames:
        #print(os.path.join(filename))
        arch_png.append(os.path.join(filename))

fig, axes = plt.subplots(2,3, figsize=(15,15)) #fila, columna, dimesion(5,5)

fil = 0
col = 0
for imagen in arch_png[0:5]:
    img = io.imread('png/'+imagen)
    #img_plot = plt.imshow(img)
    try:
        axes [fil,col].imshow(img)
        axes [fil,col].axis('off')
        axes [fil,col].set_title(imagen[0:len(imagen)-4])
        fil, col = actualizar (fil,col)
    except:
        print('Fin de muestra de 50 imagenes')
```

```python
import os
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
from skimage import io

formato = '.png'

arch_png = []
for dirname, _, filenames in os.walk('png/'):
    for filename in filenames:
        if(formato in os.path.join(filename)):
            arch_png.append(os.path.join(filename))
            #print(os.path.join(filename))

fil = 14
col = 3
fig = plt.figure(figsize = (20,20))
plt.subplots_adjust(top=10, bottom=5, left= 0.10 , hspace= 0.05,wspace= 0.025) 
#cant = len(arch_png)
print('Se muestran las imagenes creadas con el m√≥dulo APLpy')
for i in range(1,43):  #col*fil+1
    img = io.imread('png/'+arch_png[i-1])
    titulo = arch_png[i-1]
    fig.add_subplot(fil,col,i)
    plt.imshow(img)
    plt.axis('off')
    plt.title(titulo[0:len(titulo)-4])
plt.show()

arch_png2 = []
for dirname, _, filenames in os.walk('astropy_files/'):
    for filename in filenames:
        arch_png2.append(os.path.join(filename))
fil = 4
col = 2
fig = plt.figure(figsize = (20,20)) #20,20
#plt.subplots_adjust( left= 0.10)
plt.subplots_adjust(top=3, bottom=2, left= 0.1 , hspace= 0.025,wspace= 0.01) 
print('Se muestran las imagenes creados con la librer√≠a Astropy')
for i in range(1,9):  #col*fil+1
    img = io.imread('astropy_files/'+arch_png2[i-1])
    titulo = arch_png2[i-1]
    fig.add_subplot(fil,col,i)
    plt.imshow(img)
    plt.axis('off')
    plt.title(titulo[0:len(titulo)-4])
plt.show()
```
