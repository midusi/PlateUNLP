```python
#Listar imagenes
import os
import aplpy
import numpy

arch_fits = []
for dirname, _, filenames in os.walk('fits/'):
    for filename in filenames:
        #print(os.path.join(filename))
        arch_fits.append(os.path.join(filename))
```

```python
def namePNG(arch):
    long_total = len(arch)
    long_parcial = long_total - 5
    #print (arch[0:long_parcial])
    return arch[0:long_parcial]

def fits_to_png(fits):
    #To start off, use the FITSFigure class to create a canvas to where your data will be plotted:
    gc = aplpy.FITSFigure('fits/'+fits)
    #and display the image using a grayscale stretch:
    gc.show_grayscale()
    gc.tick_labels.set_font(size='x-small')
    #Sirve para mostrar el valor de cada pixel en la figura
    gc.add_colorbar()
    #Se guarda el archivo PNG generado
    nuevo = namePNG(fits)
    gc.save('png/'+nuevo+'.png')
    gc.close()
    
```

Se eligen las primeras 20 imagenes (orden alfabetico).


```python
#lista de 0..4 imagenes 
for img_fits in arch_fits[5:7]:
    #print(img_fits)
    fits_to_png(img_fits)
```

Se modifica la ejecuciÃ³n previa considerando en este caso las excepciones. De modo de convertir solo aquellas que lo permiten.

```python
def fits_to_png_v2(fits):
    #To start off, use the FITSFigure class to create a canvas to where your data will be plotted:
    try:
        gc = aplpy.FITSFigure('fits/'+fits)
        #and display the image using a grayscale stretch:
        gc.show_grayscale()
        gc.tick_labels.set_font(size='x-small')
        #Sirve para mostrar el valor de cada pixel en la figura
        gc.add_colorbar()
        #Se guarda el archivo PNG generado
        nuevo = namePNG(fits)
        gc.save('png/'+nuevo+'.png')
        list_ok.append(fits)
        gc.close()
        return True
    except OSError:
        list_error.append(fits)
        print('No se puede convertir el archivo '+fits)
        return False
    
#lista de 0..6 imagenes 
list_ok = []
list_error = []
for img_fits in arch_fits[0:7]:
    #print(img_fits)
    fits_to_png_v2(img_fits)
```

```python
print ('lista de imagenes convertidas a PNG')
for elem in list_ok:
    print (elem)
print('lista de imagenes sin conversion')
for elem in list_error:
    print (elem)
```
