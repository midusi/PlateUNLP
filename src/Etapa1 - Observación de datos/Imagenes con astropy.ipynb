```python
from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits

image_file = get_pkg_data_filename('A4518-HD93162.fits')
fits.info(image_file)
```

```python
lista_error = ['A4453-BetaMon.fits',
'A4514-HD93162.fits',
'A4518-HD93162.fits',
'C1281-HD104994.fits',
'C1281-HD115473.fits',
'C3267-29Psc.fits',
'C3267-HD693.fits',
'C3267-NGC246.fits']
print (len(lista_error))
```

```python
def namePNG(arch):
    long_total = len(arch)
    long_parcial = long_total - 5
    #print (arch[0:long_parcial])
    return arch[0:long_parcial]
#plt.figure()
for file_fits in lista_error:
    image_file = get_pkg_data_filename('fits/'+file_fits)
    image_data = fits.getdata(image_file, ext=0,ignore_missing_end=True)
    plt.figure()
    plt.imshow(image_data, cmap='gray')
    plt.colorbar()
    plt.savefig('png/astropy_files/'+namePNG(file_fits)+'.png')
```
