```python
arch = []
import os
def obtener_listado(carpeta,vector):
    for dirname, _, filenames in os.walk(carpeta):
        for filename in filenames:
            vector.append(os.path.join(filename))
    return vector
```

```python

h= pd.DataFrame({})
def agregar(carpeta):
    vector = []
    for etiqueta in cols:
        for elem in arch:
            meta_info = fits.getheader(carpeta+elem, ignore_missing_end=True) 
            vector.append( meta_info[etiqueta])
        h[etiqueta]= [dato for dato in vector]
        vector= []
        
arch=[]
arch= obtener_listado('fits/',arch)
agregar('fits/')
h
```

```python
#anios = [i for i in range(5)]

anios = []
def anios_crear(carpeta):
    for elem in arch:
        meta = fits.getheader(carpeta+elem, ignore_missing_end=True) 
        anios.append( meta['DATE-OBS'][:4])

anios_crear('fits/')

h['YEAR'] = anios

h
```
